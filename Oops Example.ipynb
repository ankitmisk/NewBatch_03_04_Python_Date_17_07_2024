```python
class ATM:
    import time
    instance  = 1
    def __init__(self,bank_name):
#         print(self)
#         print(id(self))
        self.pin = '0000'
        self.balance = 2000
        print(f'Welcome to {bank_name} Bank'.upper().center(70,' '))
        ATM.instance += 1
        
        
        pin_ans = input('Enter your PIN: ')
        
        if pin_ans == self.pin:
        
        
            ans = input('''Select any one option given below
            1>> Withdrawl
            2>> Deposit
            3>> Pin Change
            4>> Check Balance
            5>> Cancel:
            ''')


            if ans == '1':
                self.withdrawl()

            elif ans == '2':
                self.deposit()

            elif ans == '3':
                self.change_pin()

            elif ans == '4':
                self.check_balance()
            elif ans == '5':
                self.cancel()
            else:
                print('Invalid choice!!!')

        else:
            print('Invalid PIN!!!')
            exit()
    #-------------------------------------------------------Init
        
    def withdrawl(self):
        amount = input('Enter amount: ')
        if amount.isdigit():
            amount = int(amount)
            if amount<=self.balance:
                if amount % 100 == 0:
                    if amount<=50000:
                
                        print(f'Amount Rs {amount} withdrawl successfully!!')
                        self.balance -= amount
                        self.check_balance()        
                    else:
                        print('Daily Limit Exceed!!')  
                else:
                    print('Amount Must be multiple of Rs 100')
            else:
                print('Insufficient Balance!!')   
        else:
            print('Amount Must be Numerical')
        

    #----------------------------------------------------Withdrawl
    def deposit(self):
        deposit_amount = input('Enter amount: ')
        
        if deposit_amount.isdigit():
            deposit_amount = int(deposit_amount)
            if deposit_amount % 100 == 0:
                print(f'Amount Rs {deposit_amount} Deposit successfully!!')    
                self.balance += deposit_amount
                
                self.check_balance()
                
            else:
                print('Amount Must be multiple of Rs 100')    
        else:
            print('Amount Must be Numerical')
        
    
    def change_pin(self):
        import time
        pin_input = input('Enter old PIN: ')
        if pin_input == self.pin:
            new_pin_1 = input('Create New PIN: ')            
            
            if new_pin_1.isdigit() and len(new_pin_1) == 4:
                new_pin_2 = input('Enter again: ')
                
                if new_pin_1 == new_pin_2:
                    phone = input('Enter Registered Phone Number: ')
                    
                    if phone.isdigit() and len(phone) == 10:
                        import random
                        otp = random.randint(1111,9999)
                        print('OTP Sending',end = '')
                        
                        for i in range(5):
                            print('.',end = '')
                            time.sleep(1)
                            
                        display(clear=True)
                        
                        print('Your OTP is: ',otp)
                        
                        otp_ans = input('Enter OTP: ')
                        otp = str(otp)
                        if otp_ans == otp:
                            self.pin = new_pin_1
                            print('PIN chnaged successfully!!')
                            
                        else:
                            print('Incorrect OTP')
                        
                    else:
                        print('Invalid Phone Number')     
                else:
                    print('PIN Mismatched!!')         
            else:
                print('Invalid PIN!!')
        else:
            print('Incorrect PIN')
            
            
    
    def check_balance(self):
        print(f'Your Current balance is Rs {self.balance}')
    
    def cancel(self):
        print('Successfully canceled!!')
```
