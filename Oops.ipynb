```python
# udf:
class Remote:
    
    
    # methods
    def volume_plus(self):
        print(self)
        print(id(self))
        print('Volume is up by 1')
        
    def volume_down(self):
        print('Volume is down by 1')
    
    def tv_on(self):
        print('TV is on!!')
        
    def tv_off(self):
        print('TV is off!!')
```
