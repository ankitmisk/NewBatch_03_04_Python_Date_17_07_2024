```python
from abc import ABC, abstractmethod
```

```python
class Metro(ABC):
    
    @abstractmethod
    def platform(self):
        print('done1')
        
    @abstractmethod
    def track(self):
        print('done2')
        
    @abstractmethod
    def seats(self):
        print('done3')
        
```

```python
b = Metro()
```

```python
class Train(Metro):
    
    def platform(self):
        print('done1')
    
    def track(self):
        print('done2')
        
    
    def seats(self):
        print('done3')
    
    def new_method(self):
        print('done4')
```

```python
c = Train()
```

```python
c.seats()
```

```python
c.new_method()
```

```python
a = Metro()
```

```python
c.platform()
```
